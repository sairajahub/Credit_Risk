1. Null Hypothesis: Interest rate for rent ownership is greater than or equal to mortgage ownership

#The null hypothesis: The interest rate for RENT ownership is less than MORTGAGE ownership.

```python
import pandas as pd
from scipy.stats import t

# Dropping columns with more than 50% null values
threshold = 0.5
columns_to_drop = df.columns[df.isnull().mean() > threshold]
cleaned_data = df.drop(columns=columns_to_drop)

# Extract interest rate data for RENT and MORTGAGE ownership
rent_interest_rate = cleaned_data[cleaned_data['homeownership'] == 'RENT']['interest_rate']
mortgage_interest_rate = cleaned_data[cleaned_data['homeownership'] == 'MORTGAGE']['interest_rate']

# Step 1: Calculate sample means
rent_mean = rent_interest_rate.mean()
mortgage_mean = mortgage_interest_rate.mean()

# Step 2: Calculate sample standard deviations
rent_std = rent_interest_rate.std()
mortgage_std = mortgage_interest_rate.std()

# Step 3: Calculate sample sizes
rent_n = rent_interest_rate.count()
mortgage_n = mortgage_interest_rate.count()

# Step 4: Calculate pooled standard error
pooled_std = ((rent_std**2 / rent_n) + (mortgage_std**2 / mortgage_n))**0.5

# Step 5: Calculate t-statistic
t_stat = (rent_mean - mortgage_mean) / pooled_std

# Step 6: Calculate degrees of freedom
df = rent_n + mortgage_n - 2

# Step 7: Calculate p-value (one-tailed test)
p_value = t.cdf(t_stat, df)

# Display the results
print(f"Rent Mean: {rent_mean:.2f}")
print(f"Mortgage Mean: {mortgage_mean:.2f}")
print(f"Rent Std Dev: {rent_std:.2f}")
print(f"Mortgage Std Dev: {mortgage_std:.2f}")
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.5f}")

# Step 8: Decision at 95% confidence level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The interest rate for RENT ownership is less than MORTGAGE ownership.")
else:
    print("Fail to reject the null hypothesis: No evidence suggests the interest rate for RENT ownership is less than MORTGAGE ownership.")
```

2. Hypothesis Testing for Loan Amount Based on Account Type

Problem Statement:
We want to test whether the loan amount for joint accounts is greater than or equal to the loan amount for individual accounts.

Hypotheses:

Null Hypothesis: The mean loan amount for joint accounts is greater than or equal to the mean loan amount for individual accounts.

Alternative Hypothesis: The mean loan amount for joint accounts is less than the mean loan amount for individual accounts.

```python
import pandas as pd
from scipy.stats import ttest_ind, levene



# Debug: Check the unique values in the application_type column
print("Unique values in 'application_type':", df['application_type'].unique())

# Separate loan amounts based on account type
joint_loans = df[df['application_type'].str.lower() == 'joint']['loan_amount']
individual_loans = df[df['application_type'].str.lower() == 'individual']['loan_amount']

# Debug: Check sizes of both groups
print(f"Number of Joint Accounts: {joint_loans.size}")
print(f"Number of Individual Accounts: {individual_loans.size}")

# Proceed only if both groups have sufficient samples
if joint_loans.size > 1 and individual_loans.size > 1:

    # Step 2: Perform the one-tailed t-test
    t_stat, p_value = ttest_ind(joint_loans, individual_loans, alternative='less')
    print(f"T-Statistic: {t_stat}")
    print(f"P-Value: {p_value}")

    # Step 3: Decision
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: The loan amount for joint accounts is less than for individual accounts.")
    else:
        print("Fail to reject the null hypothesis: No evidence suggests the loan amount for joint accounts is less than for individual accounts.")
else:
    print("Insufficient data: One or both groups have less than 2 samples.")

```

3. Chi-Square Test: Dependency between grade and homeownership

```python
import pandas as pd
from scipy.stats import chi2_contingency


# Step 3: Create a contingency table
contingency_table = pd.crosstab(df['grade'], df['homeownership'])
print("Contingency Table:")
print(contingency_table)

# Step 4: Perform the Chi-Square test of independence
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Step 5: Display results
print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(expected)

# Step 6: Hypothesis Decision at 99% Confidence Level
alpha = 0.05  # 95% confidence level
if p_value < alpha:
    print("\nConclusion: Reject the null hypothesis.")
    print("'Grade' and 'Homeownership' are dependent.")
else:
    print("\nConclusion: Fail to reject the null hypothesis.")
    print("No evidence suggests that 'Grade' and 'Homeownership' are dependent.")
```

4. Null Hypothesis: OWN homeownership and grades A & B are independent

```python
import pandas as pd
from scipy.stats import chi2_contingency



# Step 2: Filter data for OWN homeownership and grades A & B
filtered_data = df[(df['homeownership'] == 'OWN') & (df['grade'].isin(['A', 'B']))]

# Step 3: Create a contingency table
contingency_table = pd.crosstab(filtered_data['grade'], filtered_data['homeownership'])
print("Contingency Table:")
print(contingency_table)

# Step 4: Perform the Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Step 5: Display results
print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(expected)

# Step 6: Decision
alpha = 0.05  # 95% confidence level
if p_value < alpha:
    print("\nConclusion: Reject the null hypothesis.")
    print("'OWN' homeownership and grades (A & B) are dependent.")
else:
    print("\nConclusion: Fail to reject the null hypothesis.")
    print("No evidence suggests 'OWN' homeownership and grades (A & B) are dependent.")
```

5. ANOVA Test: Loan amount across grades

```python
import pandas as pd
from scipy.stats import f_oneway



# Step 2: Check if required columns exist
if 'grade' not in df.columns or 'loan_amount' not in df.columns:
    raise ValueError("Required columns 'grade' or 'loan_amount' are missing.")

# Step 3: Prepare data for ANOVA
# Group loan_amount by grade
groups = [group['loan_amount'].values for name, group in df.groupby('grade')]

# Step 4: Perform ANOVA
f_stat, p_value = f_oneway(*groups)

# Step 5: Display results
print("ANOVA Test Results:")
print(f"F-Statistic: {f_stat}")
print(f"P-Value: {p_value}")

# Step 6: Decision
alpha = 0.05  # 95% confidence level
if p_value < alpha:
    print("\nConclusion: Reject the null hypothesis.")
    print("The means of loan amounts differ significantly across grades.")
else:
    print("\nConclusion: Fail to reject the null hypothesis.")
    print("No evidence suggests that loan amounts differ across grades.")
```
